```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
gs = matplotlib.gridspec.GridSpec(nrows=3,ncols=2,width_ratios=(0.1,0.9),height_ratios=(0.1,0.2,0.7),hspace=0.5)
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,1])
ax3 = fig.add_subplot(gs[2,1])
ax4 = fig.add_subplot(gs[2,0])
```

```python
fig = plt.figure()
gs = matplotlib.gridspec.GridSpec(nrows=3,ncols=2,width_ratios=(0.1,0.9),height_ratios=(0.1,0.2,0.7),hspace=0.7)
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,1])
ax3 = fig.add_subplot(gs[2,1])
ax4 = fig.add_subplot(gs[2,0])
gs.update(bottom=0.5)
gs2 = matplotlib.gridspec.GridSpec(nrows=1,ncols=1,top=0.3)
ax5 = fig.add_subplot(gs2[0,0])
```
